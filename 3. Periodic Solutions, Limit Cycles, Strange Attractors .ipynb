## $\textbf{3. Periodic Solutions, Limit Cycles, Strange Attractors }$

### 3.1 Introduction 

**Dynamic System**: is a system in which a function describes the **time dependence** of a point in a geometrical space. Examples include the mathematical models that describe the *swinging of a clock pendulum, the flow of water in a pipe, and the number of fish each springtime in a lake*.

**Phase Space**: a multidimensional space in which each axis corresponds to one of the coordinates required to specify the state of a physical system, all the coordinates being thus represented so that a point in the space corresponds to a state of the system.



**Limit Cycle**

There are multiple ways of analyzing **non-linear systems** in the $x-y$ plane:

* We can find the critical points and analysing how the trajectories of the system looks in the *neighborhood* of each critical point. This gives some feeling for how the other trajectories can behave, at least those which pass near enough (i.e some $\delta$) to critical points.

* Another important possibility which can influence how the trajectories look is if one of the trajectories traces out a **closed curve $C$.** If this happens, the associated solution $\textbf{x}(t)$ will be *geometrically* influenced by a point which goes round and round the curve $C$ with a certain period $T$. That is, the solution vector:
$$ \textbf{x}(t) = (x(t),y(t))$$ 

will be a pair of **periodic functions** with period $T$: 

$$ x(t+T) = x(t) \quad \quad y(t+T) = y(t) \quad \quad \forall t$$

If there *exists* a closed curve $C$, then the nearby trajectories must behave something like $C$.
The possibilities are illustrated below:
* The nearby trajectories can either spiral in (converge) towards $C$ 
* They can spiral away from $C$, or
* They can themselves be closed curves. 

If the latter case does not hold — in other words, if $C$ is an **isolated closed curve** , then $C$ is called a **limit
cycle**: $\underline{stable}$, $\underline{unstable}$, or $\underline{semi-stable}$ according to whether the nearby curves $\underline{spiral\ towards}$ $C$, $\underline{away\ from}$ $C$, or $\underline{both}$.





The most *important* kind of limit cycle are the **Stable Limit Cycle**, where the nearby curves **converge** to $C$ on both sides (i.e. from $-\infty$ and $\infty$). Periodic processes in nature can often be represented as stable limit cycles, so that great interest is attached to finding such trajectories if they exist. Unfortunately, very little is known about how to show if limit cycle exists. 

### 3.2 Showing the limit cyle exists 

The main tool which has been historically used to show that the system: 

$$   \left\{
    \begin{array}\
    x'(t) = f(x,y) \\
    y'(t) = g(x,y) \\
    \end{array}
  \right. \quad \quad (3.1)$$
has a $\underline{stable}$ limit cycle is infact the **Poincare-Bendixson Theorem**

**Theorem 3.1**: *Poincare-Bendixson Theorem*

Suppose $\boldsymbol{R}$ is the *finite region* of the plane lying between *two simple-closed curves* $\boldsymbol{D1}$ and $\boldsymbol{D2}$, and $\boldsymbol{F}$ is the *velocity vector field* for the system $(3.1)$. If:
* at each point of $\boldsymbol{D1}$ and $\boldsymbol{D2}$, the field $\boldsymbol{F}$ points toward the **interior** (towards) of $\boldsymbol{R}$, and
* $\boldsymbol{R}$ contains **no critical points**,

$\implies$ there $\underline{exists}$  a **closed trajectory $C$** lying *inside* $\boldsymbol{R}$ for the system $(3.1)$.

**Intiution** 

If we start on one of the boundary curves, the solution will enter $\boldsymbol{R}$ , since the velocity vector points into the **interior** of $\boldsymbol{R}$. As time goes on, the solution can never leave $\boldsymbol{R}$, since as it approaches a boundary curve, trying to escape from $\boldsymbol{R}$, the velocity vectors are always pointing **inwards**, forcing it to stay inside $\boldsymbol{R}$. Since the solution can never leave $\boldsymbol{R}$, the only thing it can do as $t\rightarrow \infty$ is either approach a *critical point* — but by the theorem, there are no critical points. Thus by hypothesis — it spirals (converges) in towards a closed trajectory $C$ (say). Thus there *exists* a closed trajectory $C$ inside $\boldsymbol{R}$. (It cannot be an unstable limit cycle—it must be one of the other three cases shown above.)



To be able to apply the Poincare-Bendixson theorem, one has to search the vector-field $\boldsymbol{F}$ for closed-curves $\boldsymbol{D}$ along which the velocity vectors all point towards the same side.

### 3.3 Van der Pol's Equation 

An important kind of second-order non-linear equation has the form: 

$$ x''+u(x)x'+v(x) = 0 \quad \quad (Lienard \ Equation)\quad \quad (3.2)$$

This equation could be thought of as model for a *spring-mass system* where the **damping force** $u(x)$ depends on position (for example, the mass might be moving through a viscous medium of varying density), and the **spring constant** $v(x)$ depends on how much the spring is stretched. We also allow for the possibility that $u(x) < 0$ (i.e., that there is ”**negative damping**”). So we can say that that **$u(x)<0$ for small $x$** and **$u(x)>0$ for large $x$**. The simplest expression which represents this condition is: 

$$ u(x) = \epsilon(x^2-1) \quad \quad \epsilon>0$$
Let's further assume that $v(x) = x$ i.e the spring constant is linearly proportional to the distance from a given point $x_0$. Thus overall we get an equation: 

$$ x''+\epsilon(x^2-1)x'+x = 0 \quad \quad (3.3) $$

Let 
$$ q_1 = x \\
q_2 = x' $$
This leads to:
$$   \left\{
    \begin{array}\
    q_1' = q_2 \\
    q_2'= \epsilon(1-q_1^2)q_2 - q_1 \\
    \end{array}
  \right. \quad \quad (3.4)$$

Provided that Initial Conditions are given, we can solve the set of first-order ODE $(3.4)$ using Euler's Method.  





**Theorem 3.1**: (*Poincare 1882, Bendixson 1901*)

Each *bounded* solution of a two-dimensional system

$$ q_1'=f_1(q_1,q_2),\quad \quad q_2'=f_2(q_1,q_2) $$
must:
* (i) tend to a critical point $f_1=f_2=0$ for an infinity of points $x_i\rightarrow\infty$; or
* (ii) be periodic; or 
* (iii) tend to a limit cycle 

The equation $(3.3)$ has a * **unique** non-zero periodic solution* to which all other solutions converges i.e. $(3.3)$ has a limit cycle. Intuitively, think of it as modeling a non-linear spring-mass system. When $|x|$ is large, the restoring and damping forces are large, so that $|x|$ should decrease with time. But when $|x|$ gets small, the damping becomes negative, which should make $|x|$ tend to increase with time. Thus it is plausible that the solutions should **oscillate**; that it has exactly one periodic solution is a more subtle fact (which I won't prove here).

**Note**: We can't solve *Van der Pol Equation* using the method applied to Example 3 in $2.2$. This is because $(3.4)$ is a **non-linear** equation and require a different method inorder to solve these set of equations. 

An important tool for simplifying complicated non-linear systems is that of **Steady-state approximations**. I can the first-order ODE: 

$$q_1'=\frac{q_1}{\epsilon(1-q_1^2)}$$

which can be solved using Euler's method

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

f = lambda y,e: y/(e*(1-y**2))

def Euler(X_M,x_0,y_0,h,f,e):    #x_0 and y_0 represents the intial conditions y(x_0)=y_0
    if X_M<x_0:     # makes sure that X_M>x_0 
        raise ValueError("X_M must be greater than x_0")  
        
    N = int(abs(X_M-x_0)/h)
    x_n = np.linspace(x_0,X_M,N+1)
    y_n = np.zeros(N+1,dtype = 'float64')
    y_n[0] = y_0
    for i in range(N):
        y_tilde = y_n[i]+h*f(y_n[i],e)
        y_n[i+1]=y_tilde   
    return x_n,y_n
e = 10
h=0.0001
x_n,y_n = Euler(60,0,0.1,h,f,e);
plt.plot(x_n,y_n,'k',markersize=5,label='$\epsilon$={0}'.format(e));
plt.legend()
plt.grid(True)
plt.gcf().set_size_inches(15, 2)
```

```python
'''https://www.gribblelab.org/compneuro/2_Modelling_Dynamical_Systems.html '''

from scipy.integrate import odeint

def Lorenz(state,t):                    
  y_1 = state[0]
  y_2 = state[1]       # unpack the state vector
  y_3 = state[2]

  sigma = 10.0
  r = 28.0                     # these are our constants
  b = 8.0/3.0

  y_1d = sigma * (y_2-y_1)
  y_2d = (r-y_3)*y_1 - y_2      # compute state derivatives
  y_3d = y_1*y_2 - b*y_3

  return y_1d,y_2d,y_3d      # return the state derivatives

state0 = [2.0, 3.0, 4.0]
t = np.linspace(0,30,1000)

state = odeint(Lorenz, state0, t)

# do some fancy 3D plotting
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(state[:,0],state[:,1],state[:,2])
ax.set_xlabel('$y_1$')
ax.set_ylabel('$y_2$')
ax.set_zlabel('$y_3$')
plt.gcf().set_size_inches(10.5, 8)
plt.show()
```
